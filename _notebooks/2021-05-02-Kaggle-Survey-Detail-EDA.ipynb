```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

### Reading the data

```python
df = pd.read_csv("kaggle_survey_2020_responses.csv")
```

```python
df.head()
```

```python
df = df.iloc[1:,]
```

```python
df.head()
```

#### Distribution of Age of Kagglers.

```python
df['Q1'].unique()
```

```python
plt.figure(figsize = (20,7))
df['Q1'].value_counts().sort_index().plot.bar(color = 'purple')
plt.title('Age distribution of Kaggle Users that filled the survey')
plt.xlabel('User ages')
plt.ylabel('Count')
plt.show()
```
